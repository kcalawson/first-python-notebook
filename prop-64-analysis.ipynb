## Upload and read in the propositions file

```python
props = pd.read_csv("http://first-python-notebook.readthedocs.io/_static/committees.csv")
```

## Read in the list of contributions to ballot measures

```python
contribs = pd.read_csv("http://first-python-notebook.readthedocs.io/_static/contributions.csv")
```

```python
props.head()
```

```python
props.info()
```

```python
contribs.info()
```

## Check the number of propositions' committees

```python
props.prop_name.value_counts().reset_index()
```

## Filter the Proposition 64 committees

```python
prop = props[props.prop_name == "PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE."]
```

## Merge Proposition 64 committees and contributions tabs

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

## Calculate the total* amount of disclosed donations spent on the Proposition 64

*The total we generated is not the overall total raised in the campaign, and is guaranteed to be lower than the totals reported in the media and by the campaigns.

Why?

In California, campaigns are only required to disclose the names of donors who give over $100, so our data is missing all of the donors who gave less than that amount.

```python
merged.amount.sum()
```

## Sort amounts by descending values

```python
merged.sort_values("amount", ascending=False)
```

## Count support and oppose committees donations

```python
merged.committee_position.value_counts()
```

The support committees donations dominate the oppose ones, but did they indeed raise more money?

## Create a support tab for the support committees

```python
support = merged[merged.committee_position == "SUPPORT"]
```

## Sum up the total disclosed contributions in favor of Proposition 64

```python
support.amount.sum()
```

## Create an oppose tab for the oppose committees

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

## Sum up the total disclosed contributions against Proposition 64

```python
oppose.amount.sum()
```

The support committees raised undoubtfully more money than the oppose committees. Now, let us check exactly what percent is it of the overall disclosed total.

## Percentage of support donations

```python
print support.amount.sum() / merged.amount.sum() * 100, "%"
```

## Committees that raised the most money for and against

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

## Contributors who donate the most for and against Proposition 64

```python
merged.groupby(["committee_name_x", "contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

## Create a contributors donations from outside California tab

```python
noncacontribs = merged[merged.contributor_state != "CA"]
```

## Calculate total of disclosed contributors coming from outside California

```python
noncacontribs.amount.sum()
```

## Percentage of donations from people living outside of California

```python
noncacontribs.amount.sum() / merged.amount.sum() * 100
```
